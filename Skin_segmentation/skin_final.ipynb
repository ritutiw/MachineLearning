```python
### PROBLEM STATEMENT ###
### Predict whether the sample is skin or non-skin ###

# The text file was downloaded to the location - C:\Python\Skin_Segmentaion\Skin.txt

# USER who runs this script is requested to modify the above location for both .txt and .csv file as required

#importing PANDAS
import pandas as pd

# Converting text to csv and displaying column values

print("########################")
print("Converting skin.txt file to skin.csv")


data = pd.read_csv('C:\Python\Skin_Segmentaion\Skin.txt', delimiter='\t', names=('B','G','R','TARGET'))
data.to_csv(r'C:\Python\Skin_Segmentaion\Skin.csv',index=None)

print("########################")
print("File has been converted to .CSV format")

# csv file was read using pandas library and a dataframe ( data ) was created

data = pd.read_csv (r'C:\Python\Skin_Segmentaion\Skin.csv')

# Casual check to see how the first five records are displayed in the dataframe.

data.head()
```

```python
#Loading Dependent and independent attributes into X and y respectively


X=data.iloc[:,:-1]
y=data.iloc[:,-1]

      
#splitting the data into Training and Testing set
      
from sklearn.model_selection import train_test_split
            
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

# Scaling of data using and Min and Max Scalar: Feature scaling
from sklearn.preprocessing import MinMaxScaler
      
# fit scaler on training data
scaler=MinMaxScaler()
# transform training data
X_train=scaler.fit_transform(X_train)
# transform testing dataabs
X_test=scaler.transform(X_test)
data.describe()
```

```python
#Importing and fitting the model to our dataset using Decision tree Classifier

from sklearn.tree import DecisionTreeClassifier



# Creating Decision Tree classifer object

clf=DecisionTreeClassifier(criterion="gini",splitter="best",random_state=0)

# Training Decision Tree Classifer

clf.fit(X_train,y_train)

#Predicting the response for test dataset

y_pred=clf.predict(X_test)

print(y_pred)

## Model Accuracy, how often is the classifier correct?

from sklearn import metrics

print("Accuracy of the model is :",metrics.accuracy_score(y_test, y_pred))
print("########################")
```

```python
# Building the confusion matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print("########################")
print("Confusion matrix is as below : ")
print("########################")

cm
```
